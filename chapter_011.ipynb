# 第11章 matplotlibの使い方

## 11.1 1種類のデータを可視化する

## 11.1.1 グラフにデータをプロットする

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

## 11.1.2 グラフの表示範囲を設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.ylim([0, 1])

plt.plot(x, y)
plt.show()
```

## 11.1.3 グラフの要素に名前を設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.title('y = sin(x) (0 < y < 1)')

plt.xlabel('x-axis')
plt.ylabel('y-axis')


plt.ylim([0, 1])


plt.plot(x, y)
plt.show()
```

## 11.1.4 グラフにグリッドを表示する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.title('y = sin(x)')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True)
plt.plot(x, y)
plt.show()
```

## 11.1.5 グラフの軸に目盛りを設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.title('y = sin(x)')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True)

positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]
labels = ['0°', '90°', '180°', '270°', '360°']

plt.xticks(positions, labels)

plt.plot(x, y)
plt.show()
```

## 11.2 複数のデータを可視化する 1

### 11.2.1 1つのグラフに2種類のデータをプロットする

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]
labels = ['0°', '90°', '180°', '270°', '360°']


plt.title('Graphs of trigonometric functoins')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True)

plt.xticks(positions, labels)

plt.plot(x, y1, color='k')
plt.plot(x, y2, color='b')

plt.show()
```

### 11.2.2 系列ラベルを設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]
labels = ['0°', '90°', '180°', '270°', '360°']


plt.title('Graphs of trigonometric functoins')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True)

plt.xticks(positions, labels)

plt.plot(x, y1, color='k', label='y=sin(x)')
plt.plot(x, y2, color='b', label='y=cos(x)')

plt.legend(['y=sin(x)', 'y=cos(x)'])

plt.show()

```

## 11.3 複数のデータを可視化する 2

### 11.3.1 図の大きさを設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.figure(figsize=(4, 4))

plt.plot(x, y)
plt.show()
```

### 11.3.2 サブプロットを作成する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)


fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

ax.plot(x, y)

for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```

### 11.3.3 サブプロットのまわりの余白を調整する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

labels = ['0°', '90°', '180°', '270°', '360°']
positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

plt.subplots_adjust(wspace=1, hspace=1)

ax.plot(x, y)

for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```

### 11.3.4 サブプロット内のグラフの表示範囲を設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

labels = ['0°', '90°', '180°', '270°', '360°']
positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

plt.subplots_adjust(wspace=1, hspace=1)

ax.set_ylim([0, 1])

ax.plot(x, y)

for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```

### 11.3.5 サブプロット内のグラフの要素に名前を設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

labels = ['0°', '90°', '180°', '270°', '360°']
positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]

fig = plt.figure(figsize=(9, 6))

plt.subplots_adjust(wspace=1, hspace=1)

ax = fig.add_subplot(2, 3, 5)
ax.set_title('y=sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, y)

for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```

### 11.3.6 サブプロット内のグラフにグリッドを表示する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

labels = ['0°', '90°', '180°', '270°', '360°']
positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]

fig = plt.figure(figsize=(9, 6))

plt.subplots_adjust(wspace=1, hspace=1)

ax = fig.add_subplot(2, 3, 5)

ax.grid(True)
ax.set_title('y=sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, y)

for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```

### 11.3.7 サブプロット内の軸に目盛りを設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)

y1 = np.sin(x)
y2 = np.cos(x)

labels = ['0°', '90°', '180°', '270°', '360°']
positions = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]

fig = plt.figure(figsize=(9, 6))

plt.subplots_adjust(wspace=1, hspace=1)

ax = fig.add_subplot(2, 3, 5)

ax.grid(True)
ax.set_title('y=sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_xticks(positions)
ax.set_xticklabels(labels)

ax.plot(x, y)

for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```

## 添削問題

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x_upper = np.linspace(0, 5)
x_lower = np.linspace(0, 2 * np.pi)
x_tan = np.linspace(-np.pi/2, np.pi/2)


positions_upper = [i for i in range(5)]
positions_lower = [0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2]
positions_tan = [-np.pi/2, 0, np.pi/2]

labels_upper = [i for i in range(5)]
labels_lower = ['0°', '90°', '180°', '270°', '360°']
labels_tan = ['-90°', '0°', '90°']

fig = plt.figure(figsize=(12, 9))


for i in range(6):
    ax = fig.add_subplot(2, 3, i+1)
    ax.grid(True)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    if i in [0, 1, 2]:
        ax.set_xticks(positions_upper)
        ax.set_xticklabels(labels_upper)  
    elif i in [3, 4]:
        ax.set_xticks(positions_lower)
        ax.set_xticklabels(labels_lower)        
    elif i in [5]:
        ax.set_xticks(positions_tan)
        ax.set_xticklabels(labels_tan)
        
    if i == 0:
        ax.set_title('y=x**1')
        ax.plot(x_upper, x_upper ** 1)
    elif i == 1:
        ax.set_title('y=x**2')
        ax.plot(x_upper, x_upper ** 2)
    elif i == 2:
        ax.set_title('y=x**3')
        ax.plot(x_upper, x_upper ** 3)
    elif i == 3:
        ax.set_title('y=sin(x)')
        ax.plot(x_lower, np.sin(x_lower))
    elif i == 4:
        ax.set_title('y=cox(x)')
        ax.plot(x_lower, np.cos(x_lower))
    elif i == 5:
        ax.set_title('y=tan(x)')
        ax.set_ylim(-1, 1)
        ax.plot(x_tan, np.tan(x_tan))

        
plt.show()
```
